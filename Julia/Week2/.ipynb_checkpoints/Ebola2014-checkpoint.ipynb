```julia
using DelimitedFiles
wikiEVDraw = DelimitedFiles.readdlm("wikipediaEVDraw.csv", ',')
```

```julia
using Dates
Dates.DateTime(wikiEVDraw[1,1], "d u y")
```

```julia
col1 = wikiEVDraw[:, 1]
```

```julia
for i = 1:length(col1)
    col1[i] = Dates.DateTime(col1[i], "d u y")
end
```

```julia
col1
```

```julia
Dates.datetime2rata(col1[1])
```

```julia
dayssincemar22(x) = Dates.datetime2rata(x) - Dates.datetime2rata(col1[54])
epidays = Array{Int64}(undef, 54)
for i = 1:length(col1)
    epidays[i] = dayssincemar22(col1[i])
end
```

```julia
wikiEVDraw[:, 1] = epidays # overwriting the column with epidays
DelimitedFiles.writedlm("wikipediaEVDdatesconverted.csv", wikiEVDraw, ',')
# writing the new dates to file wikipediaEVDdates..
```

```julia
using DelimitedFiles
EVDdata = DelimitedFiles.readdlm("wikipediaEVDdatesconverted.csv", ',')
epidays = EVDdata[:, 1] # all entries of column 1
allcases = EVDdata[:, 2]
```

```julia
using Plots
gr()
```

```julia
plot(epidays, allcases) #displays the graph
```

```julia
plot(epidays, allcases, linetype = :scatter, marker = :diamond)
#changing the plots
```

```julia
using DelimitedFiles
EVDdata = DelimitedFiles.readdlm("wikipediaEVDraw.csv", ',')
EVDdata[end-9:end, :]#include last 10 rows from 9
```

```julia
rows, cols = size(EVDdata) # replaced with 0 if there are any empty values
#size is function to find no of elements
for j = 1:cols
    for i = 1:rows
        if !isdigit(string(EVDdata[i, j])[1])
            EVDdata[i,j] = 0
        end
    end
end
```

```julia
EVDdata[end-9:end, :]
```

```julia
#extract data
epidays = EVDdata[:, 1]
EVDcasesbycountry = EVDdata[:, [4, 6, 8]]
# load plots and plot them
using Plots
gr()
plot(epidays, EVDcasesbycountry)
```
