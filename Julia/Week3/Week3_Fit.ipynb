```julia
using Plots
r =5

x0, y0 = 1, 2
function trialcircle(r, x0, y0)
    t = linspace(0, 2π, 200)
    x, y = r*cos(t), r*sin(t)
    x, y = x+x0, y+y0
    plot(x,y)
end
```

```julia
trialcircle(1,1,1)
#circle will display
```

```julia
function trialcircle(r, x0, y0)
    t = linspace(0, 2π, 200)
    x, y = r*cos(t), r*sin(t)
    x, y = x+x0, y+y0
    plot(x,y)
end
```

```julia
trialcircle(1, 0, 0)
trialcircle(2, 0, 0)
```

```julia
trialcircle(0, 0, 0)
for r = .1:.1:4
    trialcircle(r, 0, 0)
end
plot!()
#set of circles with grid is displayed-4,4
```

```julia
plot!(grid = false, legend = false, aspect_ratio = 1)
#modifies the circles
```

```julia
trialcircle(0, 0, 0)
for r = .1:.1:4
    trialcircle(r, r, 0)
end
plot!(grid = false, legend = false, aspect_ratio = 1)
```
